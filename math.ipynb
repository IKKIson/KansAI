## Python Basic Math Function
- 파이썬 기본 내장 함수
- Functions
  - abs : 절대값
  - max : 최댓값
  - min: 최솟값
  - pow: 멱승(제곱)
  - round: 반올림(정수)    * x.5일 경우 짝수 쪽 정수로 출력

```python
# 파이썬 기본 내장 함수
print(abs(-3))
print(max(2, 3, 5))
print(min(2, 3, 5))
print(pow(2, 3))
print(round(2.5))
print(round(3.5))
```

## Math
- python 기본 내장 모듈
- 수학 라이브러리
- Funcs - 정수 관련 함수
  - fabs : 절대값
  - ceil : 올림
  - floor : 내림
  - modf : 입력 값을 정수와 소수부분을 분리
  - factorial : 팩토리얼
  - trunc : 버림함수 (소수값 없애기)
- Funcs - 제곱/로그 관련 함수
  - exp : 지수함수 e의 x제곱
  - log : a를 b제곱근
  - sqrt : 제곱근
  - pow : x를 y제곱
- Funcs - 상수/각도 관련 함수
  - degrees : 라디안으로 표현된 각도를 60분법 각도로 변환
  - radians : 60분법으로 표현된 각도를 라디안 각도로 변환
  - pi : 원주율
  - e : 자연상수
- Funcs - 상수/각도 관련 함수
  - sin : 사인
  - cos : 코사인
  - tan : 탄젠트 
  - asin : 아크사인
  - acos : 아크코사인
  - atan : 아크탄젠트

```python
from math import *

'''
정수 관련 함수
'''
# 절대값 (결과값은 실수=float)
print(fabs(-2))
# 올림
print(ceil(2.1))
print(ceil(-2.1))
# 내림
print(floor(2.1))
print(floor(-2.1))
# 실수를 정수, 소수 분리
print(modf(3.141592))
# 팩토리얼
print(factorial(6))
# 버림함수
print(trunc(3.141592))
print(trunc(-3.141592))
print(trunc(3.741592))
print(trunc(-3.741592))
```

```python
import math
'''
제곱/로그 관련 함수
'''
# 지수함수(e^x)
print(math.exp(1))
# 로그
print(math.log(2.71828))
print(math.log(100, 2))
# 루트(제곱근)
print(math.sqrt(4.0))
# 제곱
print(math.pow(2, 3))
```

```python
import math
'''
상수/각도 관련 함수
'''
# 파이
print(math.pi)
# 자연상수
print(math.e)
# 라디안 > 도
print(math.degrees(1.57))
# 도 > 라디안
print(math.radians(90))
```

```python
from math import *
'''
삼각함수 관련 함수
'''
# 사인
print(sin(3.14159/2))
# 코사인
print(cos(3.14159/2))
# 탄젠트
print(tan(3.14159/4))
# 사인역
print(asin(1.0))
# 코사인역
print(acos(1.0))
# 탄젠트역
print(atan(3.14159))
```
